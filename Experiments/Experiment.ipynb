```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
google_api_key = os.getenv('GOOGLE_API_KEY')
```

```python
if google_api_key=="":
    print("No API key provided")
else:
    print("API key provided")
```

```python
from llama_index.llms.gemini import Gemini
import google.generativeai as genai

from llama_index.core import SimpleDirectoryReader

from llama_index.core import VectorStoreIndex
from llama_index.core import StorageContext , load_index_from_storage

from IPython.display import Markdown, display

from llama_index.embeddings.gemini import GeminiEmbedding
```

```python
genai.configure(api_key=google_api_key)
```

```python
for models in genai.list_models():
    print(models)
```

```python
for models in genai.list_models():
    if 'generateContent' in models.supported_generation_methods:
        print(models.name)
```

```python
documents = SimpleDirectoryReader("../data")
```

```python
doc = documents.load_data()
```

after loading data we now convert it to embedding and the above step is called data ingestion

search embedding models name

```python
for models in genai.list_models():
    if 'embedContent' in models.supported_generation_methods:
        print(models.name)
```

Load Model

```python
model = Gemini(model='models/gemini-3-flash-preview', api_key=google_api_key)
```

load embdding model

```python
gemini_embed_model = GeminiEmbedding(model_name="models/text-embedding-004")
```

service context use to configure similar embeddings

```python
from llama_index.core import Settings

Settings.llm = model
Settings.embed_model = gemini_embed_model
Settings.chunk_size = 800
```

now create vector from data

```python
index = VectorStoreIndex.from_documents(doc)
```

 now we will create a local data base for storing the embeddings

```python
index.storage_context.persist()
```

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("what is machine learning")
```

```python
print(response.response)
```
